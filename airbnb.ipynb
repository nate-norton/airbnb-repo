```python
jackson = pd.read_json('data/airbnb_jackson.json')


# Split the 'location' column into 'lat' and 'lon'
jackson[['lat', 'lon']] = pd.DataFrame(jackson['location'].tolist(), index=jackson.index)

# Drop the original 'location' column
jackson.drop('location', axis=1, inplace=True)

# Print the updated DataFrame
jackson.head()
```

```python
import folium

# Assuming you already have a DataFrame called 'jackson' with 'lat' and 'lon' columns

# Create a map centered around the first row's coordinates
map_center = [jackson['lat'].iloc[0], jackson['lon'].iloc[0]]
map_object = folium.Map(location=map_center, zoom_start=12)

# Iterate over each row in the DataFrame and add a marker for each location
for index, row in jackson.iterrows():
    folium.Marker(location=[row['lat'], row['lon']]).add_to(map_object)

# Save the map as an HTML file
map_file = 'map.html'
map_object.save(map_file)

from IPython.display import IFrame
IFrame(src=map_file, width=700, height=500)
```
